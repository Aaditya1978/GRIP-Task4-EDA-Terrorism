# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
%matplotlib inline
```

# Getting the Dataset

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1') # using encoding to read dataset
df.head()
```

# Getting Info of Dataset

```python
# Getting shape of the datset
df.shape
```

```python
# Getting column names
df2 = df.columns
for i in df2:
    print(i)
```

```python
# Getting information about columns like their types
df.info()
```

```python
# checking null values
df2 = df.isnull().sum()
for i in df2:
    print(i)
```

```python
# Gettting Description of the dataset
df.describe()
```

```python
# Getting number of unique values of each column
df.nunique()
```

# Getting useful Colums only

```python
#Getting the columns which has the meaningful information

df = df[['iyear','imonth','iday','country_txt','region_txt','provstate','city',
            'success','suicide','nwound','nkill','attacktype1_txt','targtype1_txt',
             'gname','natlty1_txt','weaptype1_txt']]
```

```python
df.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'country','region_txt':'region',
                     'provstate':'state','nwound':'wound','nkill':'kill','attacktype1_txt':'attack',
                     'targtype1_txt':'target','gname':'group','natlty1_txt':'nationality','weaptype1_txt':'weapon'
                    },inplace=True)
df.head()
```

# Filling Null Values

```python
df.isnull().sum()
```

```python
# Filling kill and wound column null values by mode
df.kill.fillna(df.kill.mode()[0],inplace=True)
df.wound.fillna(df.wound.mode()[0],inplace=True)
# dropping other null value columns
df.dropna(inplace=True)
```

```python
df.head()
```

# Exploratory Data Analysis

## Year Vs Success

```python
plt.figure(figsize=(20, 20))
sns.countplot(x = 'year',hue='success',data= df)
plt.title('year vs success')
plt.show()
```

## Year Vs Suicide

```python
plt.figure(figsize=(20, 20))
sns.countplot(x = 'year',hue='suicide',data= df)
plt.title('year vs suicide')
plt.show()
```

## Year Vs Kill

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='year',y='kill',data=df)
plt.title("year VS kill")
```

## Attack Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['attack'])['kill'].sum().sort_values(ascending = False).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("Attack Vs Kill")
plt.show()
```

## Country Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['country'])['kill'].sum().sort_values(ascending = False).head(15).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("country Vs Kill")
plt.show()
```

## Target Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['target'])['kill'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("target vs Kill")
plt.show()
```

## Group Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['group'])['kill'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("Group Vs Kill")
plt.show()
```

## Weapon Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['weapon'])['kill'].sum().sort_values(ascending = False).head(5).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("Weapon Vs Kill")
plt.show()
```

## Nationality Vs Kill

```python
plt.figure(figsize = (10,5))
df.groupby(['nationality'])['kill'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'red')

plt.xticks(rotation=90)
plt.title("Nationality Vs Kill")
plt.show()
```

## Attack VS Target

```python
plt.figure(figsize = (20,25))
sns.countplot(x='attack',hue='target',data = df)
plt.title('attack vs Target')
plt.show()
```

## Weapon Vs Target

```python
plt.figure(figsize = (20,25))
sns.countplot(x='weapon',hue='target',data = df)
plt.title('Weapon vs Target')
plt.show()
```
